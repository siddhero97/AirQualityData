```python
import requests

import numpy as np 
import pandas as pd 
```

```python
locationUP  = pd.read_csv("data/bigquery-public-data-openaq-global_air_quality-all.csv")
locList = locationUP["location"].tolist()
```

```python
for pollutant in ["co","no2","o3","pm10","pm25","so2"]:
    results = []
    for location in locList:
        print("location ",location," pollutant ",pollutant)
        response = requests.get("https://api.openaq.org/v1/measurements?country=IN&location="+location+"&parameter="+pollutant)
        # print(response.status_code)
        if response.status_code == 200:
            results += response.json()["results"]
    results_pd = pd.DataFrame(results)
    results_pd["year-month"]=results_pd["date"].apply(lambda x: x['local'][0:4]+"-"+x['local'][5:7])
    g = results_pd.groupby(
    ['city','location','year-month']
    ).mean()
    print(g)
#     g.to_csv("data/locationUP.csv")
```
